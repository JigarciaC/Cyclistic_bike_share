#### **Descargar, almacenar y leer los 12 archivos desde su pagina web.**

```r
# Definir la función para leer y almacenar datos
read_and_store_data <- function(url, df_name) {
  # Descargar el archivo ZIP localmente
  zip_file <- tempfile(fileext = ".zip")
  download.file(url, destfile = zip_file, mode = "wb")

  # Leer el archivo ZIP local y cargar los datos con vroom
  df <- vroom::vroom(zip_file)

  # Asignar el dataframe al nombre proporcionado
  assign(df_name, df, envir = .GlobalEnv)

  # Imprimir las primeras filas del dataframe
  cat("Primeras filas de", df_name, ":\n")
  print(head(get(df_name)))
}
```

```r
# Crear la tabla de frecuencias por tipo de bicicleta (rideable)
tabla_frecuencias_rideable <- table(trips_clear$rideable_type)

# Obtener el tipo de bicicleta uso con mayor frecuencia
tipo_bicicleta_mas_frecuente <- names(tabla_frecuencias_rideable[which.max(tabla_frecuencias_rideable)])

# Calcular el porcentaje de viajes que representa el dia más frecuente
porcentaje_bicicletas_mas_frecuente <- (tabla_frecuencias_rideable[which.max(tabla_frecuencias_rideable)] / sum(tabla_frecuencias_rideable)) * 100

# Imprimir la tabla de frecuencias y la información sobre el dia de la semana más frecuente
print(tabla_frecuencias_rideable)

cat("El tipo de bicicletas con mayor frecuencia de viajes es", tipo_bicicleta_mas_frecuente,
    "que representa un", round(porcentaje_bicicletas_mas_frecuente, 2), "% de los viajes totales, en el Sistema Cyclistic durante el periodo de estudio.\n")
```

##### **Viajes por tipo de bicicletas, por tipo de ciclista:**

```r
# Crear tabla de frecuencia. Tipo de bicicletas y tipo de ciclistas.
resultado_bicicletas_ciclistas <- trips_clear %>%
  group_by(member_casual, rideable_type) %>% # Se agrupa por tipo de usuario y tipo de bicicletas
  summarise(viajes_porcentaje = (n() / nrow(trips_clear)) * 100, .groups ="drop") %>%  # Calcula el porcentaje del total de viajes.[, .groups ="drop", porque se creó mutate antes de agrupar]
  arrange(member_casual, rideable_type) # Ordenamos el campo horas en forma descendente

# Imprimir la tabla de frecuencias por tipo y hora
print(resultado_bicicletas_ciclistas)

# Obtener la hora con mayor frecuencia de viajes para cada tipo de ciclista
tabla_frecuencias_bicicletas_ciclistas <- resultado_bicicletas_ciclistas %>%
  group_by(member_casual) %>%
  summarise(bicicleta_mas_frecuente = rideable_type[which.max(viajes_porcentaje)],
            porcentaje_viajes_mas_frecuente = max(viajes_porcentaje))

# Imprimir la información sobre la hora más frecuente por tipo de ciclista

print(tabla_frecuencias_bicicletas_ciclistas)
# Imprimir la hora de viajes con mayor frecuencia para cada tipo de ciclista
cat("La bicicleta con mayor frecuencia de uso para miembros es",
    tabla_frecuencias_bicicletas_ciclistas$bicicleta_mas_frecuente[tabla_frecuencias_bicicletas_ciclistas$member_casual == "member"],
    "y para ciclistas casuales es",
    tabla_frecuencias_bicicletas_ciclistas$bicicleta_mas_frecuente[tabla_frecuencias_bicicletas_ciclistas$member_casual == "casual"],
    "\n")

# Imprimir el porcentaje de viajes que representa la hora más frecuente para cada tipo de ciclista
cat("El porcentaje de viajes que representa el uso de la bicicleta más frecuente para miembros es",
    round(tabla_frecuencias_bicicletas_ciclistas$porcentaje_viajes_mas_frecuente[tabla_frecuencias_bicicletas_ciclistas$member_casual == "member"], 2),
    "% y para ciclistas casuales es",
    round(tabla_frecuencias_bicicletas_ciclistas$porcentaje_viajes_mas_frecuente[tabla_frecuencias_bicicletas_ciclistas$member_casual == "casual"], 2),
    "%.\n")
```

Las variables que se analiza para conocer la estaciones más utilizadas para iniciar y finalizar viajes, son respectivamente: **"start_station_name"** y **"end_station_name"**

##### **Estaciones más utilizadas para iniciar y finalizar viajes en el sistema(Cyclistic)**

```r
# Ranking de estaciones más utilizadas para iniciar viajes
ranking_start_station <- trips_clear %>%
  count(start_station_name, name = "Count", sort = TRUE) %>%
  top_n(5, Count)

# Ranking de estaciones más utilizadas para finalizar viajes
ranking_end_station <- trips_clear %>%
  count(end_station_name, name = "Count", sort = TRUE) %>%
  top_n(5, Count)

# Combinar ambos rankings en una sola tabla
combined_rankings <- bind_rows(
  ranking_start_station %>% mutate(Type = "Start"),
  ranking_end_station %>% mutate(Type = "End")
)

# Visualizar la tabla combinada
print(combined_rankings)
```

##### **Estaciones más utilizadas para iniciar y finalizar viajes, por tipo de ciclistas**

```r
# Top 5 de las estaciones con más registros de salida, por tipo de ciclista
ranking_start_station_cyclist <- trips_clear %>%
  group_by(member_casual, start_station_name) %>%
  count(sort = TRUE) %>%
  top_n(5, wt = n)

print(ranking_start_station_cyclist)
```

```r
# Top 5 de las estaciones con más registros de llegada, por tipo de ciclista
ranking_end_station_cyclist <- trips_clear %>%
  group_by(member_casual, end_station_name) %>%
  count(sort = TRUE) %>%
  top_n(5, wt = n)

print(ranking_end_station_cyclist)
```
